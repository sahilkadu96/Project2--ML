```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://www.timeanddate.com/holidays/jordan/2018"
headers = {'Accept-Language': 'en-US,en;q=0.5'}
```

```python
html = requests.get(url, headers = headers)
```

```python
soup = BeautifulSoup(html.text, 'lxml')
```

```python
table = soup.find('table', id="holidays-table")
```

```python
table_rows = table.find_all('tr', id = re.compile('^tr\d'))
```

```python
table_rows[0]
```

```python
date = []
weekday = []
holiday_name = []
holiday_type = []

for i in range(len(table_rows)):
    date.append(table_rows[i].find('th').text + ' ' + '2018')
    weekday.append(table_rows[i].find('td').text)
    holiday_name.append(table_rows[i].find('a').text)
    holiday_type.append(table_rows[i].find_all('td')[-1].text)
    
    
```

```python
df = pd.DataFrame({'Date':date, 'Weekday':weekday, 'Holiday_name': holiday_name, 'Holiday_type':holiday_type})
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.to_csv('Jordan_Holidays_2018.csv', index = False)
```
